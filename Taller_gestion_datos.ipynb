```python


datos_meta = datos[datos["Departamento"] ==  "Meta"]


datos_meta= datos_meta[['Municipio', 'Cultivo', 'Año', 'Periodo', 'Área cosechada (ha)', 'Producción (t)', "Rendimiento (t/ha)"]]
datos_meta.head()
```

```python
datos_cosecha = datos_meta.sort_values(by="Área cosechada (ha)", ascending = False)
datos_cosecha

datos_meta.to_excel("/content/drive/MyDrive/DIPLOMADO/TALLER. Modulo 1/resultados/ Area Cosechada.xlsx", index=False)
```

```python

datos_meta.isnull().sum()
```

```python
datos_meta_ajustado = datos_meta.fillna("Desconocido")
datos_meta_ajustado.isnull().sum()

datos_meta_ajustado.to_excel("/content/drive/MyDrive/DIPLOMADO/TALLER. Modulo 1/resultados/ Datos meta ajustados.xlsx", index=False)
```

```python
datos_rendi = datos_meta_ajustado.sort_values(by="Rendimiento (t/ha)", ascending = False)
datos_rendi

datos_meta_ajustado.to_excel("/content/drive/MyDrive/DIPLOMADO/TALLER. Modulo 1/resultados/ Rendimiento.xlsx", index=False)
```
